```python
from bachelors_thesis.data.preprocessing import bandpass, highpass
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Load the ECG data
signals = np.load('../../data/processed/ptbxl100all/train.npy')
signals.shape
```

```python
idx = 3
# View a signal before bandpass filtering
sig = signals[idx, :, 0]
plt.plot(sig)
```

```python
bandpass_sig = bandpass(sig, 100, order=4, low=0.5, high=40)
plt.plot(bandpass_sig)
```

```python
# highpass
highpass_sig = highpass(sig, 100, order=4, low=0.5)
plt.plot(highpass_sig)
```

```python
# Compare original, bandpass, and highpass signals
plt.figure(figsize=(12, 8))
plt.subplot(3, 1, 1)
plt.plot(sig)
plt.title('Original Signal')
plt.subplot(3, 1, 2)
plt.plot(bandpass_sig)
plt.title('Bandpass Filtered Signal')
plt.subplot(3, 1, 3)
plt.plot(highpass_sig)
plt.title('Highpass Filtered Signal')
plt.tight_layout()
plt.show()
```
